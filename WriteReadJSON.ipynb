```python
import json
lista = []
with open('rete.json') as file:
    data = json.load(file)
    for client in data['gn_retencion_anexo']:
      if client['id'] >50:
        data_tmp = {
            "model": "general.RetencionAnexo",
            "pk": client['id'],
            "fields": {
              "estado":  client['estado'],
              "fecha_registro": client['fecha_registro'],
              "usuario_registro": client['fecha_registro'],
              "codigo": client['codigo'],
              "descripcion": client['descripcion'],
              "porcentaje": client['porcentaje'],
              "tipo_retencion_anexo_id": client['tipo_retencion_anexo_id'],
              "codigo_formulario_103": client['codigo_formulario_103']
            }
        }
        lista.append(data_tmp)

json_final = {'gn_retencion_anexo': lista}

print(json_final)

with open('results_final_tildes.json', 'w', encoding="utf-8") as result_file:
    json.dump(json_final, result_file, ensure_ascii=False)



```
